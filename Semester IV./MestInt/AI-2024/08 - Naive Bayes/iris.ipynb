```python
from sklearn import datasets as ds
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB

# Step 1 - Load dataset
iris = ds.load_iris()

print("Data:")
print(iris.data)

print("Target:")
print(iris.target)
```

```python
import matplotlib.pyplot as plt

plt.title("The distribution of class labels")
plt.hist(iris.target)
plt.show()
```

```python
# Step 2 - Split dataset into training and test splits
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=0)

# Step 3 - Build the model
clf = GaussianNB()

# Step 4 - Fit the neural network to the training data
clf.fit(X_train, y_train)

# Step 5 - Interpret the results
from sklearn.metrics import accuracy_score, precision_score

y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)

print("Training ACC:", accuracy_score(y_true=y_train, y_pred=y_pred_train))
print("Test ACC:", accuracy_score(y_true=y_test, y_pred=y_pred_test))
```
